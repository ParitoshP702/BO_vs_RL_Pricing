```python
def channel_episode(policy):
  C = 10.
  S = 10.
  c,s = C,1.
  reward_array = []
  action_array = []
  state_array = []
  while s<=S or c>0:
    state_array.append([c,s])
    prob = policy(torch.tensor([c,s]))[0]
    unif = np.random.uniform(0,1)
    if unif<prob:
      price = 0.5
      param = 1-price
      un = np.random.uniform(0,1)
      action_array.append(0)
      if un<param:

        reward_array.append(price)
        # action_array.append(price)
        s+=1
        c-=1
      else:
        reward_array.append(0)
        # action_array.append(0)
        s+=1

    else:
      price = 0.7
      param = 1-price
      un = np.random.uniform(0,1)
      action_array.append(1)
      if un<param:
        reward_array.append(price)
        # action_array.append(price)
        s+=1
        c-=1
      else:
        reward_array.append(0)
        # action_array.append(0)
        s+=1


  # for i in range(0,len(returns)):
  #   return_array.append(np.sum(returns[i:]))
  return state_array,action_array,reward_array,[s,c]
```

```python
def reinforce(policy,episodes,alpha = 1e-4):
  policy_array = []
  optimizer = AdamW(policy.parameters(),lr = alpha)
  stats = []
  for episode in range(1,episodes+1):
     state_array,action_array,reward_array,_ = channel_episode(policy)
     G  = torch.zeros((1,1))
     transitions = []
     for i in range(len(reward_array)):
        transitions.append([state_array[i],action_array[i],reward_array[i]])
     for t,(state_t,action_t,reward_t) in reversed(list(enumerate(transitions))):
        G = G+reward_t
        probs_t = policy(torch.tensor(state_t))
        log_probs_t = torch.log(probs_t+1e-6)
        action_log_probs = log_probs_t[action_t]
        pg_loss = -action_log_probs*G
        total_loss_t = pg_loss.mean()
        policy.zero_grad()
        total_loss_t.backward()
        optimizer.step()
        # stats.append(total_loss_t)
     if episode < 10:
      path = "REINFROCE_bernd0"+str(episode)+".txt"
     else:
      path = "REINFROCE_bernd"+str(episode)+".txt"
     torch.save(policy,path)
  return policy
```

```python
def natural_policy_gradient(policy,episodes,epsilon = 0.1):
  for episode in range(1,episodes+1):
     state_array,action_array,reward_array,_ = channel_episode(policy)
     G  = torch.zeros((1,1))
     transitions = []
     for i in range(len(reward_array)):
        transitions.append([state_array[i],action_array[i],reward_array[i]])
     for t,(state_t,action_t,reward_t) in reversed(list(enumerate(transitions))):
        G = G+reward_t
        probs_t = policy(torch.tensor(state_t))
        w1,b1,w2,b2 = policy.parameters()
        log_probs_t = torch.log(probs_t+1e-6)
        action_log_probs = log_probs_t[action_t]
        pg_loss_for_fisher = -action_log_probs
        total_loss_for_fisher = pg_loss_for_fisher.mean()
        policy.zero_grad()
        total_loss_for_fisher.backward(gradient = torch.tensor(1))
        # print(w1.grad,b1.grad,w2.grad,b2.grad)
        # w1,b1,w2,b2 = w1.data,b1.data,w2.data,b2.data
        # print(w1,b1,w2,b2)
        w1_reshape = torch.reshape(w1.data,(-1,))
        w2_reshape = torch.reshape(w2.data,(-1,))
        theta = torch.cat((w1_reshape,b1.data,w2_reshape,b2.data))
        theta_grad = torch.cat((torch.reshape(w1.grad,(-1,)),b1.grad,torch.reshape(w2.grad,(-1,)),b2.grad))
        sz = theta_grad.numpy().shape[0]
        theta_grad = torch.reshape(theta_grad,shape = (sz,1))
        fisher_mat = torch.matmul(theta_grad,torch.t(theta_grad))
        j_theta = theta_grad*G
        print(fisher_mat)
        u,s,v = torch.linalg.svd(fisher_mat)
        tol = torch.Tensor.max(s)*1e-6
        s_inv = torch.divide(1.0,s)
        s_inv = torch.where(s<tol,0,s_inv)
        # print(u)
        # print(v)
        s_inv = torch.diag(s_inv)
        F_inv = torch.matmul(s_inv,torch.t(u))
        F_inv = torch.matmul(torch.t(v),F_inv)
        g_inv = torch.matmul(F_inv,j_theta)
        g_divide = torch.matmul(torch.t(j_theta),g_inv)
        # print(g_divide)
        alpha = torch.sqrt(torch.divide(2*epsilon,torch.matmul(torch.t(j_theta),g_inv)))
        update = alpha*g_inv
        update = torch.reshape(update,(-1,))
        print(alpha)
        theta = theta+update
        # print(theta)
        theta_np = theta.detach().numpy()
        # print(alpha)
        # print(theta_np.shape)
        w1_np,b1_np,w2_np,b2_np = theta_np[:10],theta_np[10:15],theta_np[15:25],theta_np[25:]
        w1_np = w1_np.reshape((5,2))
        w2_np = w2_np.reshape( (2,5))
        w1 = torch.tensor(w1_np)
        b1 = torch.tensor(b1_np)
        w2 = torch.tensor(w2)
        b2 = torch.tensor(b2)
        diction = {"0.weight":w1,"0.bias":b1,"2.weight":w2,"2.bias":b2}
        policy.load_state_dict(diction,strict = False)
     if episode < 10:
      path = "NPG_bernd0"+str(episode)+".txt"
     else:
      path = "NPG_bernd"+str(episode)+".txt"
     torch.save(policy,path)

  return policy
```

```python
critic_hidden = 32
critic = nn.Sequential(nn.Linear(2,critic_hidden),
                       nn.ReLU(),
                       nn.Linear(critic_hidden,1),
                       nn.ReLU())

```

```python
from torch.optim import Adam
critic_optimizer = Adam(critic.parameters(),lr = 0.005)
def update_critic(advantages):
  loss = .5*(advantages**2).mean()
  critic_optimizer.zero_grad()
  loss.backward()
  critic_optimizer.step()
```

```python
def estimate_advantages(states,last_state,rewards):
  values = critic(states)
  last_value = critic(last_state.unsqueeze(0))

  next_values = torch.zeros_like(rewards)
  for i in reversed(range(rewards.shape[0])):
    last_value = next_values[i] = rewards[i]+last_value
  advantages  = next_values - values
  return advantages
```

```python
def surrogate_loss(new_probabilities,old_probabilities,advantages):
  return (new_probabilities/old_probabilities* advantages).mean()
```

```python
def kl_div(p,q):
  p = p.detach()
  return (p*(p.log()-q.log())).sum(-1).mean()
```

```python
def flat_grad(y,x,retain_graph = False,create_graph = False):
  if create_graph:
    retain_graph = True

  g = torch.autograd.grad(y,x,retain_graph = retain_graph,create_graph = create_graph)
  g = torch.cat([t.view(-1) for t in g])
  return g
```

```python
def conjugate_gradient(A,b,delta =1e-5,max_iterations = 1000):
  x = torch.zeros_like(b)
  r = b.clone()
  p = b.clone()

  i = 0
  while i<max_iterations:
    AVP = A(p)
    dot_old = r@r
    alpha = dot_old/(p@AVP)
    x_new = x+alpha*p

    if (x-x_new).norm()<=delta:
      return x_new
    i+=1
    r = r-alpha*AVP

    beta = (r@r)/dot_old
    p = r+beta*p

    x = x_new
  return x
```

```python
def update_agent(rollouts,actor):
  states = torch.cat([r.states for r in rollouts],dim = 0)
  actions = torch.cat([r.actions for r in rollouts],dim = 0).flatten()

  advantages = [estimate_advantages(states,next_states[-1],rewards) for states,_,rewards,next_states in rollouts]
  advantages = torch.cat(advantages,dim = 0).flatten()


  advantages = (advantages - advantages.mean())/advantages.std()
  update_critic(advantages)

  distribution= actor(states)
  distribution = torch.distributions.utils.clamp_probs(distribution)
  probabilities = distribution[range(distribution.shape[0]),actions]


  L = surrogate_loss(probabilities,probabilities.detach(),advantages)
  KL = kl_div(distribution,distribution)


  parameters = list(actor.parameters())
  g = flat_grad(L,parameters,retain_graph = True)
  d_kl = flat_grad(KL,parameters,create_graph = True)
  # print(g)
  # print(d_kl)
  def HVP(v):
    return flat_grad(d_kl@v,parameters,retain_graph = True)
  # print(torch.autograd.functional.jacobian(d_kl,parameters))
  search_dir = conjugate_gradient(HVP,g)
  # search_dir = scipy.sparse.linalg.cg()
  # print(HVP(g))
  # print(d_kl@g)

  delta = 0.1
  # print(HVP(search_dir))
  # print(torch.linalg.det(HVP(search_dir)))
  # print(g@search_dir)
  val = search_dir@HVP(search_dir)
  if val<0:
     val = torch.tensor(1e-4)
  print(val)
  max_length  = torch.sqrt(2*delta/val)
  print(max_length)
  max_step = max_length*search_dir
  # print("Hey there!!!!!")
  def criterion(step):
    apply_update(step)

    with torch.no_grad():
      distribution_new = actor(states)
      distribution_new = torch.distributions.utils.clamp_probs(distribution_new)
      probabilities_new = distribution_new[range(distribution_new.shape[0]),actions]

      L_new = surrogate_loss(probabilities_new,probabilities,advantages)
      KL_new = kl_div(distribution,distribution_new)
    L_improvement = L_new-L
    if L_improvement > 0 and KL_new<=delta:
      print(f"Yes...updated!!!!!")
      # print(actor.state_dict())
      return True
    apply_update(-step)
    return False
  i = 0
  while not criterion((0.9**i)*max_step) and i<10:
    i+=1
  return actor

```

```python
def surrogate_loss(new_probabilities,old_probabilities,advantages):
  return (new_probabilities/old_probabilities* advantages).mean()
```

```python
def kl_div(p,q):
  p = p.detach()
  return (p*(p.log()-q.log())).sum(-1).mean()
```
